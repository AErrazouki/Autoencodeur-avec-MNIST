# Importation des Bibliothèques

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

# Chargement et Prétraitement des données

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Afficher les dimensions des ensembles de données
print("Dimensions originales de x_train:", x_train.shape)
print("Dimensions originales de x_test:", x_test.shape)
```

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
# Aplatir les images (28x28 pixels) en vecteurs de taille 784
x_train = x_train.reshape((x_train.shape[0], -1))
x_test = x_test.reshape((x_test.shape[0], -1))
```

```python
# Définir la taille de l'entrée
input_dim = x_train.shape[1]
print("Dimension d'entrée :", input_dim)
```

# Construction du modèle d'autoencodeur

```python
def build_deep_autoencoder(input_dim):
    # Encodeur
    input_layer = layers.Input(shape=(input_dim,))
    encoded = layers.Dense(512, activation='relu')(input_layer)  
    encoded = layers.Dense(256, activation='relu')(encoded)       
    encoded = layers.Dense(128, activation='relu')(encoded)       
    bottleneck = layers.Dense(64, activation='relu')(encoded) 
    
    # Décodeur
    decoded = layers.Dense(128, activation='relu')(bottleneck)    
    decoded = layers.Dense(256, activation='relu')(decoded)       
    decoded = layers.Dense(512, activation='relu')(decoded)       
    output_layer = layers.Dense(input_dim, activation='sigmoid')(decoded) 
    
    # Définir le modèle d'autoencodeur
    autoencoder = models.Model(input_layer, output_layer)
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
    return autoencoder
```

```python
# Construire l'autoencodeur
autoencoder = build_deep_autoencoder(input_dim)
```

# Entraînement du Modèle

```python
# Entraîner l'autoencodeur
history = autoencoder.fit(x_train, x_train, epochs=10, batch_size=256, shuffle=True, validation_data=(x_test, x_test))
```

# Évaluation et Visualisation

```python
# Évaluer l'autoencodeur
reconstructed = autoencoder.predict(x_test)
```

```python
# Visualiser la perte d'entraînement
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Visualiser les images originales et reconstruites

```python
def visualize_reconstruction(x_original, x_reconstructed, num_images=10):
    plt.figure(figsize=(20, 4))
    for i in range(num_images):
        # Afficher les images originales
        ax = plt.subplot(2, num_images, i + 1)
        plt.imshow(x_original[i].reshape(28, 28), cmap='gray')
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        # Afficher les images reconstruites
        ax = plt.subplot(2, num_images, i + 1 + num_images)
        plt.imshow(x_reconstructed[i].reshape(28, 28), cmap='gray')
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()

# Visualiser les premières images originales et reconstruites
visualize_reconstruction(x_test, reconstructed)
```
